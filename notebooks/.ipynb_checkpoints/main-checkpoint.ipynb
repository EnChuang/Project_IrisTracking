```python
import cv2
import os
import mediapipe as mp
import numpy as np
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import time
import winsound
import subprocess
```

```python
# MediaPipe Tasks
BaseOptions = python.BaseOptions
FaceLandmarker = vision.FaceLandmarker
FaceLandmarkerOptions = vision.FaceLandmarkerOptions
VisionRunningMode = vision.RunningMode

# 模型檔路徑
model_path = "face_landmarker.task"
# 警示影片路徑（改成你實際位置）
ALERT_VIDEO_PATH = r"SkeletonShield.mp4"
#POTPLAYER_PATH = r"C:/Program Files/DAUM/PotPlayer/PotPlayerMini64.exe"

base_options = BaseOptions(
    model_asset_path=model_path
)

options = FaceLandmarkerOptions(
    base_options=base_options,
    running_mode=VisionRunningMode.VIDEO,  # 用 VIDEO 模式處理連續畫面
    num_faces=1,
    min_face_detection_confidence=0.5,
    min_face_presence_confidence=0.5,
    min_tracking_confidence=0.5,
    output_face_blendshapes=False,  # 不需要表情，可關閉
    output_facial_transformation_matrixes=False  # 不需要轉換矩陣
)

face_landmarker = FaceLandmarker.create_from_options(options)

# 虹膜 + 眼睛 landmarks 索引
LEFT_IRIS = [474, 475, 476, 477]   # 左眼虹膜（常用 474 作為中心）
RIGHT_IRIS = [469, 470, 471, 472]  # 右眼虹膜（常用 469 作為中心）

LEFT_EYE = [362, 382, 381, 380, 374, 373, 390, 249, 263, 466, 388, 387, 386, 385, 384, 398]
RIGHT_EYE = [33, 7, 163, 144, 145, 153, 154, 155, 133, 173, 157, 158, 159, 160, 161, 246]
```

```python
cap = cv2.VideoCapture(0)

start_time = time.time() * 1000

# 沒看螢幕的計時變數
no_look_start_time = None
NO_LOOK_REMIND_SEC = 0.6
remind_sound_played = False

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        print("Can't access the cam")
        break

    current_timestamp_ms = int((time.time() - start_time) * 1000)

    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    detection_result = face_landmarker.detect_for_video(mp_image, current_timestamp_ms)

    if detection_result.face_landmarks:
        for landmarks in detection_result.face_landmarks:
            h, w, _ = frame.shape

            # 畫左虹膜輪廓
            # left_iris_points = [[int(landmarks[i].x * w), int(landmarks[i].y * h)] for i in LEFT_IRIS]
            # if left_iris_points:
            #     pts = np.array(left_iris_points, np.int32)
            #     cv2.polylines(frame, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

            # 畫右虹膜輪廓
            right_iris_points = [[int(landmarks[i].x * w), int(landmarks[i].y * h)] for i in RIGHT_IRIS]
            # if right_iris_points:
            #     pts = np.array(right_iris_points, np.int32)
            #     cv2.polylines(frame, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

            # 虹膜中心點
            left_iris_pt = (int(landmarks[474].x * w), int(landmarks[474].y * h))
            right_iris_pt = (int(landmarks[469].x * w), int(landmarks[469].y * h))
            # cv2.circle(frame, left_iris_pt, 4, (0, 255, 255), -1)
            # cv2.circle(frame, right_iris_pt, 4, (255, 255, 0), -1)

            # 眼部中心
            left_eye_center_x = sum(landmarks[i].x for i in LEFT_EYE) / len(LEFT_EYE)
            left_eye_center_y = sum(landmarks[i].y for i in LEFT_EYE) / len(LEFT_EYE)
            left_eye_center = (int(left_eye_center_x * w), int(left_eye_center_y * h))

            right_eye_center_x = sum(landmarks[i].x for i in RIGHT_EYE) / len(RIGHT_EYE)
            right_eye_center_y = sum(landmarks[i].y for i in RIGHT_EYE) / len(RIGHT_EYE)
            right_eye_center = (int(right_eye_center_x * w), int(right_eye_center_y * h))

            # 左眼眼皮垂直距離（上眼皮最高點 - 下眼皮最低點）
            left_eye_top_y = min(landmarks[i].y for i in LEFT_EYE)
            left_eye_bottom_y = max(landmarks[i].y for i in LEFT_EYE)
            left_eye_open_height = (left_eye_bottom_y - left_eye_top_y) * h

            # 右眼眼皮垂直距離
            right_eye_top_y = min(landmarks[i].y for i in RIGHT_EYE)
            right_eye_bottom_y = max(landmarks[i].y for i in RIGHT_EYE)
            right_eye_open_height = (right_eye_bottom_y - right_eye_top_y) * h

            # 用眼眶寬度標準化（避免遠近影響）
            left_eye_width = (max(landmarks[i].x for i in LEFT_EYE) - min(landmarks[i].x for i in LEFT_EYE)) * w
            right_eye_width = (max(landmarks[i].x for i in RIGHT_EYE) - min(landmarks[i].x for i in RIGHT_EYE)) * w

            # 開合比例（高度 / 寬度，越大越開）
            left_open_ratio = left_eye_open_height / left_eye_width if left_eye_width > 0 else 0
            right_open_ratio = right_eye_open_height / right_eye_width if right_eye_width > 0 else 0

            avg_open_ratio = (left_open_ratio + right_open_ratio) / 2

            # 顯示 debug
            # print(f"左眼開合比例: {left_open_ratio:.4f}  右眼: {right_open_ratio:.4f}  平均: {avg_open_ratio:.4f}")
            # cv2.putText(frame, f"Open Ratio: {avg_open_ratio:.3f}", (50, 120), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 0), 2)

            OPEN_THRESHOLD = 0.31  
            is_looking = avg_open_ratio > OPEN_THRESHOLD

            # 狀態顯示
            status = "Looking" if is_looking else "Not Looking"
            color = (0, 255, 0) if is_looking else (0, 0, 255)
            #cv2.putText(frame, status, (50, 80), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3, cv2.LINE_AA)

            # 沒看持續時間
            current_time = time.time()
            if not is_looking:
                if no_look_start_time is None:
                    no_look_start_time = current_time
                    remind_sound_played = False

                if current_time - no_look_start_time >= NO_LOOK_REMIND_SEC:
                    cv2.putText(frame, "Focus Alarm", (100, 110), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 4, cv2.LINE_AA)
                    cv2.rectangle(frame, (0, 0), (w, h), (0, 0, 255), thickness=10)
                    
                    if not remind_sound_played:
                        os.startfile(ALERT_VIDEO_PATH)          
                        remind_sound_played = True
            else:
                no_look_start_time = None
                remind_sound_played = False

            # 畫眼部中心
            #cv2.circle(frame, left_eye_center, 2, (0, 0, 255), -1)
            #cv2.circle(frame, right_eye_center, 2, (0, 0, 255), -1)

            # debug 文字
            # cv2.putText(frame, f"L rel max: {max_left_rel:.3f}", (50, 240), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
            # cv2.putText(frame, f"R rel max: {max_right_rel:.3f}", (50, 270), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)

    else:
        # 沒偵測到臉時顯示提示
        cv2.putText(frame, "No Face Detected", (w//2 - 150, h//2), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3)

    cv2.imshow('Iris Tracking - Press q to quit', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
